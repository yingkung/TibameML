```python
from sklearn.datasets import load_boston
boston = load_boston()  # load 数据集
boston
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['target'] = boston['target']
df.to_csv('boston.csv', encoding='utf-8', index=False)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,8))
sns.heatmap(df.astype('float').corr(),cmap= 'YlOrBr', annot=True)
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
# train_test_split --> (训练特征90%， 测试特征10%， 训练目标 90%， 测试目标10%)
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), df['target'], test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(reg, out_file=None, feature_names=boston['feature_names'],  filled=True, special_characters=True)
graph = graphviz.Source(g)
graph
```

```python
pred = reg.predict(x_test)
print('预测结果：', list(pred))
print('真正标签：', list(y_test))
```

```python
# from sklearn.metrics import accuracy_score
# print('预测成功几率：', accuracy_score(pred, y_test))
from sklearn.metrics import r2_score
print('r2 score:', r2_score(list(y_test), list(pred)))
```
