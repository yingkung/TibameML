```python
from sklearn.datasets import load_iris
iris= load_iris()
iris
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df['target'] = iris['target']
df.to_csv('iris.csv', encoding='utf-8', index=False)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,8))
sns.heatmap(df.astype('float').corr(),cmap= 'YlOrBr', annot=True)
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
# train_test_split --> (训练特征90%， 测试特征10%， 训练目标 90%， 测试目标10%)
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), df['target'], test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4)
clf.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(clf, out_file=None, feature_names=iris['feature_names'], class_names=iris['target_names'], filled=True, special_characters=True)
graph = graphviz.Source(g)
graph
```

```python
pred = clf.predict(x_test)
print('预测结果：', list(pred))
print('真正标签：', list(y_test))
```

```python
from sklearn.metrics import accuracy_score
print('预测成功几率：', accuracy_score(pred, y_test))
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,pred))  # 生成混淆矩阵
```
