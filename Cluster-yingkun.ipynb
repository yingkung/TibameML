```python
from sklearn.datasets import load_iris
iris= load_iris()
iris
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df['target'] = iris['target']
df.to_csv('iris.csv', encoding='utf-8', index=False)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,8))
sns.heatmap(df.astype('float').corr(),cmap= 'YlOrBr', annot=True)
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
# åˆ‡åˆ†æ•°æ®
# train_test_split --> (è®­ç»ƒç‰¹å¾90%ï¼Œ æµ‹è¯•ç‰¹å¾10%ï¼Œ è®­ç»ƒç›®æ ‡ 90%ï¼Œ æµ‹è¯•ç›®æ ‡10%)
x_train, x_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), df['target'], test_size=0.1)
```

```python
# åšå…¬å¼ï¼ˆfitï¼‰
# åœ¨å·²çŸ¥ç¾¤æ•°æ—¶ï¼Œ K å°±ç›´æŽ¥é€‰
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(x_train)
```

```python
pred = clu.predict(x_test)
print('é¢„æµ‹ç»“æžœï¼š', list(pred))
print('çœŸæ­£æ ‡ç­¾ï¼š', list(y_test))
```

```python
from sklearn.metrics import silhouette_score
x = df.drop(['target'], axis=1)
plotx = []
ploty = []
for k in range(2,6):
    clu = KMeans(n_clusters=k)
    clu.fit(x)
    s = silhouette_score(x, clu.labels_)
    plotx.append(k)
    ploty.append(s)
print('kå€¼ = ', plotx)
print('score=', ploty)
```
